# Tuning QGS hyper-parameters

____
### This Notebook allows you to compute probability distributions to measure marked states  according to QGS hyper-parameters.  

In the periodicity_1 module two functions are available: 

1. plot_distribution(): it computes the probability distributions of the n_marked_states as function of the number of iterations of quantum amplitude amplification (in a range [0, n_g_max]),  when $\beta$ is fixed.<br>
plot_distribution parameters: <br>
    1.1 n_qubits: number of qubits in the quantum register implementing QGS. (set as the length of bit-string individuals)<br>
    1.2 n_marked_states: number of marked states<br>
    1.3 n_g_max: upper limit of the x axis<br>
    1.4 beta: value of beta ( its upper limit is $\frac{\pi}{\text{n_marked_states+1}}$ )<br>


2. plot_distribution_vs_beta() : it computes the probability distributions of the n_marked_states as function of $\beta$ with a fixed number n_g of iterations of quantum amplitude amplification.<br> 
plot_distribution_vs_beta parameters: <br>
    2.1 n_qubits: number of qubits in the quantum register implementing QGS. (set as the length of bit-string individuals)<br>
    2.2 n_marked_states: number of marked states<br>
    2.3 n_g: number of iterations of quantum amplitude amplification process<br>
    2.4 beta_max: highest value of beta to consider (its upper limit is $\frac{\pi}{\text{n_marked_states+1}}$)<br>
    2.5 iterations_ : number of beta values to consider starting from the value beta_max e progressively dividing it for two. 

```python
import periodicity_1
import math
```

An example of plot_distribution execution with five qubits:

```python
periodicity_1.plot_distribution(n_qubits=5, n_marked_states=3, n_g_max=10, beta=math.pi/4)
```

An example of plot_distribution_vs_beta execution with five qubits:

```python
periodicity_1.plot_distribution_vs_beta(n_qubits=5, n_marked_states=3, n_g=2, beta_max=math.pi/4, iteration_=4)
```
